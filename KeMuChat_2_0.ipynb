# Running the application

```python
# Run Flask in the background
import subprocess
import time

flask_process = subprocess.Popen(['python', 'run.py'], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
```

```python
# Set up LocalTunnel
lt_process = subprocess.Popen(['lt', '--port', '5000', '--local-host', 'localhost'], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
```

```python
# Wait for LocalTunnel to start
time.sleep(15)
```

### Use this IP address in the link below as your *password.*

```python
# Fetch the tunnel password or token from the given URL for localtunnel authentication or configuration
!curl https://loca.lt/mytunnelpassword
```

```python
# Capture the LocalTunnel URL
while True:
    output = lt_process.stdout.readline()
    if output == b'' and lt_process.poll() is not None:
        break
    if output:
        decoded_output = output.strip().decode('utf-8')
        if 'https://' in decoded_output:
            public_url = decoded_output.split(' ')[-1]
            print(f"Your app is available at: {public_url}")
            break
```
